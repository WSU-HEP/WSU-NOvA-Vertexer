```python
#data_dir='/home/m962g264/research_repo/data-preprocess/MC_Nominal_FHC_testdata_abdul'
data_dir='/home/c479r989/research_repo/pv-finder-zak/trainer-model/'
```

```python
data_qe_dir=data_dir+'y_model_predictions_qe.csv'
data_res_dir=data_dir+'y_model_predictions_res.csv'
data_dis_dir=data_dir+'y_model_predictions_dis.csv'
data_cc_dir=data_dir+'y_model_predictions_cc.csv'
data_nc_dir=data_dir+'y_model_predictions_nc.csv'
```

```python
reco_qe_dir=data_dir+'y_reco_qe.csv'
reco_res_dir=data_dir+'y_reco_res.csv'
reco_dis_dir=data_dir+'y_reco_dis.csv'
reco_cc_dir=data_dir+'y_reco_cc.csv'
reco_nc_dir=data_dir+'y_reco_nc.csv'
```

```python
reco_qe=pd.read_csv(reco_qe_dir,header=None)
reco_qe.columns=['True Y','Reco Y']
reco_res=pd.read_csv(reco_res_dir,header=None)
reco_res.columns=['True Y','Reco Y']
reco_dis=pd.read_csv(reco_dis_dir,header=None)
reco_dis.columns=['True Y','Reco Y']
reco_cc=pd.read_csv(reco_cc_dir,header=None)
reco_cc.columns=['True Y','Reco Y']
reco_nc=pd.read_csv(reco_nc_dir,header=None)
reco_nc.columns=['True Y','Reco Y']
reco_full=pd.concat([reco_cc,reco_nc],axis=0)
reco_full.columns=['True Y','Reco Y']
```

```python
data_qe=pd.read_csv(data_qe_dir,header=None)
data_qe.columns=['True Y','Model Predictions','Reco Y']
data_res=pd.read_csv(data_res_dir,header=None)
data_res.columns=['True Y','Model Predictions','Reco Y']
data_dis=pd.read_csv(data_dis_dir,header=None)
data_dis.columns=['True Y','Model Predictions','Reco Y']
data_cc=pd.read_csv(data_cc_dir,header=None)
data_cc.columns=['True Y','Model Predictions','Reco Y']
data_nc=pd.read_csv(data_nc_dir,header=None)
data_nc.columns=['True Y','Model Predictions','Reco Y']
data_full=pd.concat([data_cc,data_nc],axis=0)
data_full.columns=['True Y','Model Predictions','Reco Y']
```

```python
#quasi-elastic
plt.figure(figsize=(8,8))
true_vertex=data_qe['True Y']
plt.xlabel('True Y Vertex (cm)')
plt.ylabel('Model Y Vertex (cm)')
sns.kdeplot(data=data_qe,x='True Y',y='Model Predictions',fill=True)
plt.ylim(-250,250)
plt.xlim(-250,250)
plt.plot(true_vertex,true_vertex,'r',lw=1.0,ls='--')

#calculating mean and stuff
averagex=np.mean(data_qe)
#Calculating RMS
rmsx=np.sqrt(np.mean(data_qe**2))
print(rmsx)
print(averagex)
```
